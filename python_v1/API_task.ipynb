```python
import requests


url = "https://catfact.ninja/facts"

# Send a GET request to the API
response = requests.get(url)
response.raise_for_status()  # Raise an HTTPError for bad responses (4xx and 5xx)

# Parse the JSON response
print(response)
data = response.json()
print(data)
fact = data.get("fact")
```

```python
import pandas as pd

ddf = data['data']

df = pd.DataFrame(ddf)

df.head()
```

```python
cats = df.loc[:5,:]
cats.head(15)
```

```python
cats['len'] = df['fact'].str.len()

cats['words'] = df['fact'].apply(lambda x: len(x.split()) )

cats.head()
```

```python
import numpy as np

cats['score'] = np.nan
for i, f in cats.iterrows():
    print(f['fact'])
    score = int(input(f'Podaj ocene  (1-10):'))
    cats.at[i, 'score'] = score
cats.head()    
```

```python
import random

cats['ocena x'] = np.random.randint(0, 33, size=len(cats)).astype('int')
cats['ocena y'] = np.random.randint(0, 100, size=len(cats)).astype('int')

cats.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scal1 = MinMaxScaler(feature_range=(0,10))
cats['skal ocena x'] = scal1.fit_transform(cats[['ocena x']])

scat2 = MinMaxScaler(feature_range=(0,100))

cats['skal ocena y'] = scat2.fit_transform(cats[['ocena y']])
cats.head()
```

```python
cats['sum'] = (cats['score'] + cats['ocena x'] + cats['ocena y'])
cats['avg_score'] = (cats['score'] + cats['ocena x'] + cats['ocena y'])/3
print(cats['avg_score'].max())
cats.head()
```

```python
top1_score = cats.sort_values('score', ascending=False)
top1_score.head(1)
```
